### BeautifulSoup

#### 1. http://quotes.toscrape.com

```python
# Solicitud del html a la página web de interés
import requests

url = 'http://quotes.toscrape.com'
page = requests.get(url)

print('Status Code 200: La solicitud ha sido Confirmada')
print('Status Code 404: No se ha encontrado la página web')
print('Status Code 500: Se ha producido algún error interno')
print()
print('Resultado de la solicitud:', page.status_code)
```

```python
# Creación de objeto BeautifulSoup
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content)

# Visualización del código html de forma ordenada
print(soup.prettify())
```

```python
# Extracción de las frases
quotes = soup.find_all('div',class_='quote') # Para encontrar las las "class quote" en todas las "div"
citas = []

for quote in quotes:
    cita = quote.find('span', class_ ='text').text
    citas.append(cita)

citas
```

```python
# Extracción de los autores
autores=[]

for quote in quotes:
    autor=quote.find('small', class_='author').text
    autores.append(autor)

autores
```

```python
#Extracción de los Tags
tags=[]
for quote in quotes:
    tag=quote.find('div',class_='tags').meta['content']
    tags.append(tag)
    
tags
```

```python
# Creación de un dataframe
import pandas as pd
data = {'Cita': citas, 'Autor': autores, 'Etiquetas': tags}
frases_autores = pd.DataFrame(data)

frases_autores
```

#### 2. https://en.wikipedia.org/wiki/List_of_compositions_by_Dmitri_Shostakovich

```python
# Solicitud del html a la página web de interés

page_2 = requests.get('https://en.wikipedia.org/wiki/List_of_compositions_by_Dmitri_Shostakovich')

print('Status Code 200: La solicitud ha sido Confirmada')
print('Status Code 404: No se ha encontrado la página web')
print('Status Code 500: Se ha producido algún error interno')
print()
print('Resultado de la solicitud:', page_2.status_code)
```

```python
# Creación de objeto BeautifulSoup

soup_2 = BeautifulSoup(page_2.content)


# print(soup_2.prettify() ----> Para visualización del código html de forma ordenada (no ejecutado para evitar extensión)
```

```python
# Búsqueda de la tabla que contiene la información de interés

tabla_wik = soup_2.find('table', class_ = 'wikitable')
```

```python
# Selección de las columnas de interés con sus correspondientes registros o filas
Title = []
Instrumentation = []
Year = []

filas = tabla_wik.find_all("tr")[1:]  

for fila in filas:
    fils_tabl = fila.find_all("td")
    title = fils_tabl[1].text.strip()
    instruments = fils_tabl[2].text.strip()
    year = fils_tabl[3].text.strip()

    Title.append(title)
    Instrumentation.append(instruments)
    Year.append(year)
```

```python
# Creación del dataframe con columnas de interés

data = {'Título': Title, 'Instrumentación': Instrumentation, 'Año': Year}
compositions_by_Shostakovich = pd.DataFrame(data)
compositions_by_Shostakovich

```

Se ha realizado una extracción de la información relativa a las composiciones de Dmitri Shostakovich, uno de los grandes compositores del S. XX. La tabla contiene ahora solo el "título", "instrumentación" y el "año".

### Selenium

#### 1. http://quotes.toscrape.com

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# Iniciación del controlador y acceso a la url
driver = webdriver.Chrome()
driver.get('http://quotes.toscrape.com/')
```

```python
# Extracción de las frases
citas_slnium = driver.find_elements(By.CLASS_NAME, 'text')

for cita in citas_slnium:
    cita = cita.text
    print(cita)
```

```python
# Extracción de los autores
autores_slnium = driver.find_elements(By.CLASS_NAME, 'author')

for autor in autores_slnium:
    autor= autor.text
    print(autor)
```

```python
# Extracción de los tags

tags_slnium = driver.find_elements(By.CLASS_NAME, "tags")

for tag in tags_slnium:
    tag= tag.text
    print(tag)
```

```python
# Creación de un dataframe
data_slnium = {'Cita': [cita.text for cita in citas_slnium],
               'Autor': [autor.text for autor in autores_slnium],
               'Tags': [tag.text for tag in tags_slnium]} 

frases_autores_slnium = pd.DataFrame(data_slnium) 
frases_autores_slnium['Tags'] = frases_autores_slnium['Tags'].str.replace('Tags: ', '')

frases_autores_slnium     
```

```python
# Cierre del controlador de Selenium
driver.quit()
```

### Selenium para extraer tabla de cotización de divisas en https://es.investing.com/currencies/exchange-rates-table

```python
# Iniciación del controlador y acceso a la url
driver_divisas = webdriver.Chrome()
driver_divisas.get('https://es.investing.com/currencies/exchange-rates-table')
```

```python
# tabla_divisas = driver_divisas.find_element(By.CSS_SELECTOR, '.divisas_europa')
```

```python
# Búsqueda de la tabla que contiene la información de interés

tabla_divisas = driver_divisas.find_element(By.CSS_SELECTOR,  '.genTbl')
```

```python
# Selección de las columnas de interés con sus correspondientes registros o filas
Par = []
DOLAR= []
EURO = []
LIBRA = []

filas_divisas = tabla_divisas.find_elements(By.TAG_NAME, "tr")[1:]

for fila in filas_divisas:
    fils_tabl_divisas = fila.find_elements(By.TAG_NAME, "td")
    par = fils_tabl_divisas[0].text.strip()
    dolar = fils_tabl_divisas[1].text.strip()
    euro = fils_tabl_divisas[2].text.strip()
    libra = fils_tabl_divisas[3].text.strip()

    Par.append(par)
    DOLAR.append(dolar)
    EURO.append(euro)
    LIBRA.append(libra)
```

```python
# Creación de un dataframe
data_divisas = {'Euro': EURO,
               'Dólar': DOLAR,
               'Libra': LIBRA} 

divisas = pd.DataFrame(data_divisas, index=Par) 

divisas
```

```python
# Cierre del controlador de Selenium
driver_divisas.quit()
```

La tabla muestra las cotizaciones de las divisas del Euro, Dolar y la Libra esterlina con respecto a otras divisas a fecha del 27 de septiembre del 2023 y hora 10:15, ya que estamos ante la extracción de información de una tabla dinámica.
